```python
#dieu khien dong co AC
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctie
setpoint = ctie.Antecedent(np.arange(18,28,1),'setpoint')
val_sensor = ctie.Antecedent(np.arange(0,41,1),'val_sensor')
pwm = ctie.Consequent(np.arange(0,256,1),'pwm')


setpoint['low'] = fuzz.trimf(setpoint.universe,[18,18,22])
setpoint['normal'] = fuzz.trimf(setpoint.universe,[18,22,26])
setpoint['high'] = fuzz.trimf(setpoint.universe,[22,26,27])

val_sensor['low'] = fuzz.trimf(val_sensor.universe,[0,0,17])
val_sensor['normal'] = fuzz.trimf(val_sensor.universe,[0,17,36])
val_sensor['high'] = fuzz.trimf(val_sensor.universe,[17,36,40])

pwm['low'] = fuzz.trimf(pwm.universe,[0,0,125])
pwm['normal'] = fuzz.trimf(pwm.universe,[0,125,205])
pwm['high'] = fuzz.trimf(pwm.universe,[125,205,255])

rule1 = ctie.Rule(setpoint['low']&val_sensor['low'],pwm['low'])
rule2 = ctie.Rule(setpoint['low']&val_sensor['normal'],pwm['low'])
rule3 = ctie.Rule(setpoint['normal']&val_sensor['normal'],pwm['normal'])
rule4 = ctie.Rule(setpoint['normal']&val_sensor['high'],pwm['normal'])
rule5 = ctie.Rule(setpoint['normal']&val_sensor['low'],pwm['normal'])
rule6 = ctie.Rule(setpoint['high']&val_sensor['low'],pwm['normal'])
rule7 = ctie.Rule(setpoint['high']&val_sensor['normal'],pwm['normal'])
rule8 = ctie.Rule(setpoint['low']&val_sensor['high'],pwm['normal'])
rule9 = ctie.Rule(setpoint['high']&val_sensor['high'],pwm['high'])

setpoint.view()
val_sensor.view()
pwm.view()

pwm_ctie= ctie.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
pwming = ctie.ControlSystemSimulation(pwm_ctie)

pwming.input['setpoint']=22
pwming.input['val_sensor']=17
pwming.compute()
print(pwming.output['pwm'])
pwm.view(sim=pwming)
```

```python
#dieu khien module robot
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctie
L = ctie.Antecedent(np.arange(0,151,1),'L')
C = ctie.Antecedent(np.arange(0,151,1),'C')
R = ctie.Antecedent(np.arange(0,151,1),'R')

LC = ctie.Consequent(np.arange(0,91,1),'LC')
RC = ctie.Consequent(np.arange(-90,1,1),'RC')

pwm_L = ctie.Consequent(np.arange(0,91,1),'pwm_L')
pwm_R = ctie.Consequent(np.arange(-90,1,1),'pwm_R')

L['N'] = fuzz.trimf(L.universe,[0,0,75])
L['M'] = fuzz.trimf(L.universe,[0,75,150])
L['F'] = fuzz.trimf(L.universe,[75,150,150])

C['N'] = fuzz.trimf(C.universe,[0,0,75])
C['M'] = fuzz.trimf(C.universe,[0,75,150])
C['F'] = fuzz.trimf(C.universe,[75,150,150])

R['N'] = fuzz.trimf(R.universe,[0,0,75])
R['M'] = fuzz.trimf(R.universe,[0,75,150])
R['F'] = fuzz.trimf(R.universe,[75,150,150])

RC['B'] = fuzz.trimf(RC.universe,[-90,-90,-45])
RC['N'] = fuzz.trimf(RC.universe,[-90,-45,0])
RC['S'] = fuzz.trimf(RC.universe,[-45,0,0])

LC['S'] = fuzz.trimf(LC.universe,[0,0,45])
LC['N'] = fuzz.trimf(LC.universe,[0,45,90])
LC['B'] = fuzz.trimf(LC.universe,[45,90,90])

pwm_R ['B'] = fuzz.trimf(pwm_R .universe,[-90,-90,-45])
pwm_R ['N'] = fuzz.trimf(pwm_R .universe,[-90,-45,0])
pwm_R ['S'] = fuzz.trimf(pwm_R .universe,[-45,0,0])

pwm_L['S'] = fuzz.trimf(pwm_L.universe,[0,0,45])
pwm_L['N'] = fuzz.trimf(pwm_L.universe,[0,45,90])
pwm_L['B'] = fuzz.trimf(pwm_L.universe,[45,90,90])

L.view()
R.view()
C.view()
LC.view()
RC.view()
pwm_R.view()
pwm_L.view()

rule1 = ctie.Rule(L['N']&C['N'],LC['S'])
rule2 = ctie.Rule(L['N']&C['M'],LC['N'])
rule3 = ctie.Rule(L['N']&C['F'],LC['B'])
rule4 = ctie.Rule(L['M']&C['N'],LC['S'])
rule5 = ctie.Rule(L['M']&C['M'],LC['N'])
rule6 = ctie.Rule(L['M']&C['F'],LC['N'])
rule7 = ctie.Rule(L['F']&C['N'],LC['S'])
rule8 = ctie.Rule(L['F']&C['M'],LC['N'])
rule9 = ctie.Rule(L['F']&C['F'],LC['B'])

rule10 = ctie.Rule(R['N']&C['N'],RC['S'])
rule11 = ctie.Rule(R['N']&C['M'],RC['N'])
rule12 = ctie.Rule(R['N']&C['F'],RC['B'])
rule13 = ctie.Rule(R['M']&C['N'],RC['S'])
rule14 = ctie.Rule(R['M']&C['M'],RC['N'])
rule15 = ctie.Rule(R['M']&C['F'],RC['N'])
rule16 = ctie.Rule(R['F']&C['N'],RC['S'])
rule17 = ctie.Rule(R['F']&C['M'],RC['N'])
rule18 = ctie.Rule(R['F']&C['F'],RC['B'])

rule19 = ctie.Rule(RC['S']&LC['S'],pwm_R['S'])
rule20 = ctie.Rule(RC['N']&LC['S'],pwm_R['S'])
rule21 = ctie.Rule(RC['B']&LC['S'],pwm_R['N'])
rule22 = ctie.Rule(RC['S']&LC['N'],pwm_R['S'])
rule23 = ctie.Rule(RC['N']&LC['N'],pwm_R['N'])
rule24 = ctie.Rule(RC['B']&LC['N'],pwm_R['N'])
rule25 = ctie.Rule(RC['S']&LC['B'],pwm_R['N'])
rule26 = ctie.Rule(RC['N']&LC['B'],pwm_R['N'])
rule27 = ctie.Rule(RC['B']&LC['B'],pwm_R['B'])

rule28 = ctie.Rule(RC['S']&LC['S'],pwm_L['S'])
rule29 = ctie.Rule(RC['N']&LC['S'],pwm_L['S'])
rule30 = ctie.Rule(RC['B']&LC['S'],pwm_L['N'])
rule31 = ctie.Rule(RC['S']&LC['N'],pwm_L['N'])
rule32 = ctie.Rule(RC['N']&LC['N'],pwm_L['N'])
rule33 = ctie.Rule(RC['B']&LC['N'],pwm_L['B'])
rule34 = ctie.Rule(RC['S']&LC['B'],pwm_L['N'])
rule35 = ctie.Rule(RC['N']&LC['B'],pwm_L['B'])
rule36 = ctie.Rule(RC['B']&LC['B'],pwm_L['B'])

pwm_ctie= ctie.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,
                              rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,
                              rule18,rule19,rule20,rule21,rule22,rule23,rule25,rule26,
                              rule28,rule29,rule30,rule31,rule32,rule33,rule34,rule35,rule36])
pwming = ctie.ControlSystemSimulation(pwm_ctie)

pwming.input['L']=50
pwming.input['C']=100
pwming.input['R']=100
pwming.compute()
print(pwming.output['pwm_L'])
print(pwming.output['pwm_R'])
pwm_L.view(sim=pwming)
pwm_R.view(sim=pwming)
```
